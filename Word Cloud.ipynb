```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def join_data(X):
    joined_data = " ".join([data for data in X])
    return joined_data

def tfidf_vectorizer(X):
    vectorizer = TfidfVectorizer(sublinear_tf=True)
    vectorized_X = vectorizer.fit_transform(X)
    return vectorized_X, vectorizer

def create_wordcloud(df, label):
    df = df[df.y == label]
    X = join_data(df.X)
    X, vectorizer = tfidf_vectorizer(df.X)
    f_names = vectorizer.get_feature_names()
    dense = X.todense()
    denselist = dense.tolist()
    df = pd.DataFrame(denselist, columns=f_names)
    data = df.transpose()
    return data
    #wordcloud = WordCloud().generate_from_frequencies(data)
    #plt.imshow(wordcloud)
```
